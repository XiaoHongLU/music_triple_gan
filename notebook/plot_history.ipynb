```python
import os
from keras.models import Sequential
from keras.layers.recurrent import LSTM
from keras.layers import Dense, Bidirectional
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
import numpy as np
import json
%matplotlib inline
from matplotlib import pyplot as plt

batch_size = 1000
nb_epochs = 50
X = np.load('../Input/inputX.npy')
Y = np.load('../Label/inputy.npy')
data_train_valid, data_test, label_train_valid, label_test = train_test_split(X, Y, test_size=0.2, random_state=42)
data_train, data_valid, label_train, label_valid = train_test_split(data_train_valid, label_train_valid, test_size=0.3,
                                                                    random_state=42)
```

```python
tag = 'layer_1_unit_100'
history = json.loads(open("../Models/"+tag + "/history.json","r").read())
epoch_num = 300
start_point = 3
f, axarr = plt.subplots(1,2,figsize=(16,8))

s = [["loss","val_loss"],["acc","val_acc"]]
legend_label = [["loss","val_loss"],["acc","val_acc"]]
title = ["1","2"]
for i in range(2):
    for t in s[i]:
        axarr[i].plot(range(start_point ,epoch_num),history[t][start_point:])
        axarr[i].legend(legend_label[i])
        axarr[i].set_title(title[i])
```
